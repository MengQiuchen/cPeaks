## intra two OC clusters

```r
# change back to working with peaks instead of gene activities
DefaultAssay(seurat.cpeak.atac) <- 'peaks'

da_OC_intra <- FindMarkers(group.by = 'peaks_snn_res.0.3',
  object = seurat.cpeak.atac,
  ident.1 = 0,
  ident.2 = 1,
  test.use = 'LR',
  #latent.vars = 'peak_region_fragments'
)
```

```r
qsave(da_OC_intra,'p8.da_OC_intra.qs')
```

## inter OC cluster

```r
tmp <- qread('metadata_trans.qs')

metadata.expand <- seurat.cpeak.atac@meta.data%>%rownames_to_column('a')%>%mutate(clusterID=str_extract(predictedGroup_ArchR,'.*(?=-)')%>%as.numeric,
                                clusterType=str_extract(predictedGroup_ArchR,'(?<=-).*'))%>%
                        left_join(tmp,by=c('clusterID'))%>%column_to_rownames('a')

seurat.cpeak.atac@meta.data <- metadata.expand
```

```r
# change back to working with peaks instead of gene activities
DefaultAssay(seurat.cpeak.atac) <- 'peaks'

da_OC_inter <- FindMarkers(group.by = 'clusterID',
  object = seurat.cpeak.atac,
  ident.1 = 10,
  #ident.2 = 1,
  test.use = 'LR',
  #latent.vars = 'peak_region_fragments'
)
```

```r
qsave(da_OC_inter,'p8.da_OC_inter.qs')
```
