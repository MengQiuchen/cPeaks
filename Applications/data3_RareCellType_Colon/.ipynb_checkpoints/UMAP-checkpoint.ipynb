```r
set.seed(0)

suppressPackageStartupMessages({
    library(SummarizedExperiment)
    library(Matrix)
    library(data.table)
    library(magrittr)
    library(viridis)
    library(Signac)
    library(Seurat)
    library(GenomeInfoDb)
    library(EnsDb.Hsapiens.v86)
})


fun_markers <- function(mat){
    
    chrom_assay <- CreateChromatinAssay(
                  counts = mat,
                 metadata=metadata,
                  sep = c(":", "-"),
                  genome = 'hg38',
                  # fragments = "/data1/lichen/code/ECA2.0/scATAC/data/annotation/PBMC//atac_pbmc_5k_nextgem_fragments.tsv.gz",
                  min.cells = 0,
                  min.features = 20
                )
    seurat.object <- CreateSeuratObject(
                      counts = chrom_assay,
                      assay = "peaks",
                      #meta.data = metadata
                    )


    seurat.object <- RunTFIDF(seurat.object)
    seurat.object <- FindTopFeatures(seurat.object, min.cutoff = 'q0')
    seurat.object <- RunSVD(seurat.object)

    DepthCor(seurat.object)%>%print
    
    labels <- (metadata%>%as.data.frame%>%column_to_rownames('index'))[seurat.object@meta.data%>%rownames,'cell type']
    seurat.object@meta.data=cbind(seurat.object@meta.data,labels)

    DefaultAssay(seurat.object) <- 'peaks'
    Idents(object = seurat.object) <- seurat.object@meta.data$labels

    suppressMessages({
        # seurat.object <- RunUMAP(object = seurat.object, reduction = 'lsi', dims = 2:30)
        seurat.object <- RunUMAP(object = seurat.object, reduction = 'lsi',n.neighbors = 20,min.dist = 0.1,seed.use=333,matric='euclidean',
                                 dims = 2:30)
    })
    p1 = DimPlot(object = seurat.object, label = TRUE) +
        scale_color_manual(values = colors_)+
        scale_alpha_manual(values = 0.5)+ NoLegend()


    mat.markers <- FindAllMarkers(object = seurat.object)

    seurat.object <- FindNeighbors(object = seurat.object, reduction = 'lsi', dims = 2:30)
    seurat.object <- FindClusters(object = seurat.object, verbose = FALSE, algorithm = 3)
    p2=DimPlot(object = seurat.object, label = TRUE) + NoLegend()

    mat.markers.cluster <- FindAllMarkers(object = seurat.object)
    
    return(list(seurat.object=seurat.object,
                mat.markers=mat.markers,
                mat.markers.cluster=mat.markers.cluster,
                p1=p1,p2=p2))

    
}
```

```r
cres <- qread('data/cpeaks.qs')
peaks <- qread('data/bulkPeaks.qs')
cpeaks.sub <- qread('data/cpeaks.qs')
```

```r
metadata=qread('data/metadata.qs')
labels=metadata$`cell type`
```

```r

peaks <- peaks[,metadata$index]
cres <- cres[,metadata$index]
```

```r
res.list <- c()
```

```r
mat.list <- list(cpeaks.sub =cpeaks.sub,peaks=peaks,cres=cres)
```

```r
res.list <- (mat.list)%>%mclapply(fun_markers,mc.cores=3)
```

```r
names(res.list) <- names(mat.list)
```
